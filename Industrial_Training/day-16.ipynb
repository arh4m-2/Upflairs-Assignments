#relplot()

## PIE CHART

```python
df['day'].value_counts()
```

```python
labels = list(df['day'].value_counts().keys())
labels
```

```python
values = list(df['day'].value_counts().values)
values
```

```python
plt.pie(values, labels=labels)
plt.show()
```

```python
# ec -> edgecolor , k-> black
plt.pie(values, labels=labels, wedgeprops={'ec' : 'k'} )  # wedgeprop takes a dictionary
plt.show() 
```

```python
lst = [0,0,0,0.4]  # we can change the walue of 0 to increase spacing of a slice. ( take value between 0-1)
plt.pie(values, labels=labels, wedgeprops={'ec' : 'k'}, explode=lst )
plt.show() 
```

```python
lst = [0,0,0,0.4] 
sns.set(rc={'figure.figsize': (4,4)}) # to change the size of the chart
plt.pie(values, labels=labels, wedgeprops={'ec' : 'k'}, explode=lst, autopct='%1.2f%%', shadow=True)  # .2 means upto 2 decimal places
plt.title('Cafe Data - Daywise')
# plt.savefig("PieChart")  # to save the pie chart
plt.show() 
```
