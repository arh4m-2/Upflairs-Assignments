# SCATTER CHART

Scatter Chart is use to analyse relation between x-axis values and y-axis values

```python
lst1 = [2,5,4,3,5,8,7]
lst2 = [3,4,7,9,6,3,2]
plt.scatter(lst1, lst2)   # lst1 as x-axis  and lst2 as y-axis
```

## BAR CHART

While using bar chart, always remember to use Catagotical values on x-axis and Numerical values on y axis

```python
df['owner'].value_counts()
```

```python
labels = list(df['owner'].value_counts().keys())
labels
```

```python
values = list(df['owner'].value_counts().values)
values
```

```python
plt.bar(labels,values)
plt.xticks(rotation = 60)  # rotates the names by 60
plt.show()  # shows only chart
```

```python
df['brand'].value_counts()
```

```python
labels = list(df['brand'].value_counts().keys())
```

```python
values = list(df['brand'].value_counts().values)
```

```python
plt.bar(labels, values)
plt.xticks(rotation = 60)
plt.show()
```
