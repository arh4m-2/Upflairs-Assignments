Slicing of table

```python
# 2 methods are used -> loc()  and iloc()

df.loc[0:10]  # inclusive of 0 and 10
```

```python
# sclicing for columns

df.loc[0:10, ['price', 'age']]
```

```python
# iloc in inclusive of endpoint

df.iloc[0:10]
```
