```python
from sklearn.model_selection import train_test_split 
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression  # This algorithm is used for classification problem statement
```

```python
lg = LogisticRegression()
```

```python
lg.fit(x_train,y_train)
```

```python
round(lg.score(x_train,y_train),2)*100
```

```python
round(lg.score(x_test, y_test, ),2)*100
```

OVERFIT PROBLEM -> When the score in training is higher than the score in testing by a considerable amount

High score in training -->  98%
Low score in testibng --> 75%






UNDERFIT PROBLEM -> Training sccore is low, but testing score can be either low or high

```python
pred = lg.predict(x_test)
```

```python
y_test['Prediction'] = pred
```


True Positive -> the actual value is 1(True) and the predicted value is also 1 (True)


False Positive -> the actual value is 0(False) but the predicted value is 1(True)

False Negative -> the actual value is 1(False) but the predicted value is 0(True)

True Negative -> the actual value is 0(False) and the predicted value is also 0(False)z

The True Positive and True Negative values should be high. Their high values signifies the accuracy of the model

The False Neagative and False Positive values are the error points. Thus their values should be low

Accuracy = (TP+TN) / (TP+TN+FP+FN)

Precision = TP/(TP+FP)
