Write a program to find the minimum price of ktm bike,
which are available in Jaipur, and how many ktm bikes are
available at Jaipur location.


```python
import pandas as pd
```

```python
df = pd.read_csv('Used_Bikes.csv')
```

```python
df
```

```python
ktm_data = df[(df['city']=='Jaipur') & (df['brand']=='KTM')]
ktm_data
```

```python
min_price = ktm_data['price'].min()
```

```python
min_price
```

```python
no_of_bikes = ktm_data.shape[0]
no_of_bikes
```
